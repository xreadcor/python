```python
movie = pd.read_csv('movie.csv')
print(movie.columns)
movie[['movieId', 'title']]
```

```python
rating = pd.read_csv('rating.csv')
print(rating.columns)
rating[['userId', 'movieId', 'rating']]
```

```python
data = pd.merge(movie, rating)
```

```python
grouped_users_by_rating = rating.groupby('userId').count()['rating'].to_frame()
```

```python
print('Среднее кол-во голосов от пользователей: ', grouped_users_by_rating['rating'].mean())
print('Медианное кол-во голосов от пользователей: ', grouped_users_by_rating['rating'].median())
```

```python
grouped_movies_by_rating = data.groupby('title').count()['rating'].sort_values(ascending=False).to_frame()
grouped_movies_by_rating
```

```python
print('Среднее кол-во голосов от пользователей: ', grouped_movies_by_rating['rating'].mean())
print('Медианное кол-во голосов от пользователей: ', grouped_movies_by_rating['rating'].median())
```

```python
mean_rating = pd.DataFrame(data.groupby('title').mean()['rating'])
mean_rating['number of ratings'] = pd.DataFrame(data.groupby('title').count()['rating'])
mean_rating.head(10)
```

```python
mean_rating.sort_values(by='rating', ascending=False)
```
